```python
import pandas as pd
```

```python
league_21_to_22 = pd.read_csv('top5_leagues_standings.csv').drop(columns=['league_id', 'team_id'])
league_23_to_25 = pd.read_csv('all_leagues_team_records.csv').drop(columns=['league_code', 'team_id'])
```

```python
# only include the top 5 leagues
league_23_to_25 = league_23_to_25[league_23_to_25['league_name'].isin(['Bundesliga', 'Ligue 1', 'Primera Division', 'Premier League', 'Serie A'])]
league_23_to_25['league_name'].unique()
```

```python
# rename 'Primera Division' to 'La Liga' to match other df
league_23_to_25['league_name'] = league_23_to_25['league_name'].apply(lambda x: 'La Liga' if x == 'Primera Division' else x)

# calculate goals_diff column to match other df
league_23_to_25['goal_diff'] = league_23_to_25['goals_for'] - league_23_to_25['goals_against']

# rename column names to match df
league_23_to_25 = league_23_to_25.rename(columns={'league_name': 'league'})

league_23_to_25
```

```python
# reorder columns
league_23_to_25 = league_23_to_25[['season', 'league', 'team_name', 'played', 'wins', 'draws', 'losses', 'points', 'goal_diff']]
league_23_to_25
```
